# **Importing required packages**

```python
from flask import Flask, request, render_template, url_for, jsonify
from diffusers import StableDiffusionPipeline
import torch
import os
```

# **Loading the model if it is already download else download the model**

```python
#project will download into "/content/drive/MyDrive/project/model_cache". Make sure you have enough space in your google drive
MODEL_DIR = "/content/drive/MyDrive/project/model_cache"
os.makedirs(MODEL_DIR, exist_ok=True)
try:
  # Load the CompVis/stable-diffusion-v1-4 model
  pipe = StableDiffusionPipeline.from_pretrained(
      "CompVis/stable-diffusion-v1-4",
      torch_dtype=torch.float16,
      cache_dir=MODEL_DIR
  )
  pipe.to("cuda")

  print("Model loaded successfully and ready for inference!")
except Exception as e:
    print(f"Error loading model: {e}")
```

# **Running flask app**

```python
#make sure you have every folder  in right path
app = Flask(__name__, template_folder='/content/drive/MyDrive/project/template',static_folder= '/content/drive/MyDrive/project/static')

# Folder to save generated images
OUTPUT_FOLDER = "/content/drive/MyDrive/project/static/generated-images"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

@app.route('/', methods=['GET', 'POST'])
def home():
    image_url = None
    prompt = None
    if request.method == 'POST':
        # Get the prompt from the form
        prompt = request.form.get('prompt')
        if prompt:
            # Generate the image using the prompt
            filename = generate_image(prompt)
            image_url = url_for('static', filename=f"generated-images/{filename}")

    return render_template('main-index.html', image_url=image_url, prompt=prompt)

def generate_image(prompt):
    # Generate image using the already-loaded pipe object, in this case "CompVis/stable-diffusion-v1-4" this model
    image = pipe(prompt=prompt, num_inference_steps=200, guidance_scale=5).images[0]



    # Save image with limited file name
    filename = f"{prompt[:50].replace(' ', '_')}.png"
    image_path = os.path.join(OUTPUT_FOLDER, filename)
    image.save(image_path)

    return filename

if __name__ == '__main__':
    # Get the URL for accessing the app
    from google.colab.output import eval_js
    print(eval_js("google.colab.kernel.proxyPort(5000)"))

    # Run the Flask app
    app.run(host='0.0.0.0', port=5000)
```
